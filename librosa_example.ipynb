## Librosa Library Example
 - 이 노트북은 `librosa` 라이브러리에 대한 설명을 진행합니다.
 - 이 라이브러리는 음성 및 신호처리를 위한 대표적인 파이썬 라이브러리 입니다.
 - 각 모델의 입력을 위한 특징 추출이 가능합니다.
 - 대표적인 음성 특징은 Mel-Spectrogram, MFCC, MFCC delta & deltadelta가 있습니다.

```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
```

### 0. 오디오(훈련) 데이터 경로 지정

```python
audio_path = './2501ml_data/train/KDF_T_2006.wav'
```

### 1. librosa 라이브러리를 통해 오디오 데이터 처리

```python
# audio 정보 처리
y, sr = librosa.load(audio_path, sr=None)

# 파형 시각화
plt.figure(figsize=(16, 4))
librosa.display.waveshow(y, sr=sr)
plt.title('Waveform of Example Sample')
plt.show()
```

```python
# Mel-Sepcgrogram 시각화
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
S_dB = librosa.power_to_db(S, ref=np.max)

plt.figure(figsize=(16, 4))
librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel', fmax=8000)
plt.title('Mel-frequency spectrogram of Example Sample')
plt.show()
```

```python
# MFCC 시각화
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)
plt.figure(figsize=(16, 4))
librosa.display.specshow(mfcc, sr=sr, x_axis='time')
plt.title('MFCC of Example Sample')
plt.show()
```
